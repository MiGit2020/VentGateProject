### Wind Speed vs P2 mean , Categorized by East Ventgate

p2mean_8ms.csv contain east Ventagate < 8m/s

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el archivo CSV en un DataFrame
df_p2mean = pd.read_csv('p2mean_8ms.csv')

# Crear una nueva columna para categorizar los rangos de eastVent
def categorizar_eastVent(eastVent):
    if eastVent <= 30:
        return '<=30'
    elif 30 < eastVent <= 60:
        return '30-60'
    elif eastVent <= 100:
        return '60-100'
    else:
        return '>100'  # Para cubrir cualquier valor fuera del rango

df_p2mean['EastVent_Range'] = df_p2mean['eastVent'].apply(categorizar_eastVent)

# Graficar windSpeed versus p2mean, coloreando según los rangos de eastVent
sns.lmplot(x='windSpeed', y='p2_mean', data=df_p2mean, hue='EastVent_Range', fit_reg=False, height=8, aspect=2.0)

# Añadir título y etiquetas
plt.title('Gráfico de windSpeed vs p2mean con Rangos de eastVent')
plt.xlabel('windSpeed (m/s)')
plt.ylabel('p2mean')

# Mostrar el gráfico
plt.show()

```

```python
#Next plot
#Se carga el archivo p2mean_8ms.csv en un DataFrame df utilizando pd.read_csv
#Se define la función categorizar_vent que clasifica las filas en función de los valores de eastVent y westVent.
#Si ambos valores están en el rango de 0 a 10, la categoría es 'Ambos VentGates: 0-10'.
#Si solo eastVent está en ese rango, la categoría es 'eastVent: 0-10'.
#Si solo westVent está en ese rango, la categoría es 'westVent: 0-10'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('p2mean_8ms.csv')

# Crear una nueva columna para categorizar los valores de eastVent y westVent
def categorizar_vent(eastVent, westVent):
    if 0 < eastVent <= 10 and 0 < westVent <= 10:
        return 'Ambos VentGates: 0-10'
    elif 0 < eastVent <= 10:
        return 'eastVent: 0-10'
    elif 0 < westVent <= 10:
        return 'westVent: 0-10'
    #else:
        #return 'Fuera de rango'

df['VentGates_Categoria'] = df.apply(lambda row: categorizar_vent(row['eastVent'], row['westVent']), axis=1)

# Crear el scatter plot usando seaborn
plt.figure(figsize=(15, 7))
sns.scatterplot(x='windSpeed', y='p2_mean', hue='VentGates_Categoria', data=df, palette='coolwarm')

# Añadir título y etiquetas
plt.title('windSpeed vs p2_mean with VentGates Categories')
plt.xlabel('windSpeed (m/s)')
plt.ylabel('p2_mean')

# Mostrar el gráfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('p2mean_8ms.csv')

# Filtrar el DataFrame para cumplir con las condiciones:
# eastVent: (0-10], westVent: (0-10] , 270 < windDir <= 360, az:[90-180]
df_filtrado = df[(df['eastVent'] > 0) & (df['eastVent'] <= 10) &
                 (df['westVent'] > 0) & (df['westVent'] <= 10) &
                 (df['windDir'] >= 270 ) & (df['windDir'] <= 360) &
                (df['ecs_az'] >= 90) & (df['ecs_az'] <= 180)]

print(df_filtrado[['windDir']].describe())
# Crear un scatter plot usando seaborn
plt.figure(figsize=(15, 6))
sns.scatterplot(x='windSpeed', y='p2_mean', hue='windDir', data=df_filtrado, palette='coolwarm')

# Fijar los límites del eje y
plt.ylim(0, 2)  # Ajusta los valores según tu rango deseado

# Añadir título y etiquetas
plt.title('windSpeed vs p2_mean, east Ventgate:(0-10] ,west Ventgate:(0-10], wind direction:[270-360] and ecs_az: [90-180]')
plt.xlabel('windSpeed (m/s)')
plt.ylabel('p2_mean')

# Mostrar el gráfico
plt.show()
```
